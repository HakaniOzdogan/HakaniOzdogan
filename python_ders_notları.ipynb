```python
liste = [5, 2, 8, 1, 7]

# Len: Listenin uzunluğunu(eleman sayısını) döndürür
uzunluk = len(liste)
print("Uzunluk", uzunluk)

# Max: Listenin en büyük elemanını döndürür
maximum = max(liste)
print("En büyük eleman", maximum)

# Min: Listenin en küçük elemanına döndürür
minimum = min(liste)
print("En küçük eleman", minimum)

# Sum: Listedeki elemanların toplamını dödürür
toplam = sum(liste)
print("Listenin toplamı", toplam)

# Sorted: Listenin sıralanmış bir örtüsünü döndürür, orjinal liste değişmez
sirali_liste = sorted(liste)
print("Sıralı liste :", sirali_liste)

# Index: Belirtilen bir elemaını döndürür (ilk bulduğu indexi)
index = liste.index(8)  # Eğer(8,4) yazılırsa 4. indexten aramayı başlatır
print("Elemanın indexi :", index)

# Count: Belirtilen bir elemanın iste içerisinde kaç defa tekrar ettiği yazılır
sayi = liste.count(5)
print("Tekrar etme sayısı: ",sayi)

# Append: Liste sonuna bir eleman ekler
liste.append(4)
print("Yeni liste(Append) :", liste)

# Pop: Belirtilen indexteki elemanı kaldırır
pop_eleman = liste.pop(2)
print("Kaldırılan eleman(Pop): ", pop_eleman)
print("Yeni liste(Pop): ", liste)

# Del: Belirtilen indexteki elemanı kaldırır
del liste[0]
print("Yeni Liste (Del): ", liste)

#Reverse: Belirtilen listeyi tersine çevirir
liste.reverse()
print("Yeni Liste(Reverse): ", liste)

#Sort: Listeyi sıralar
liste.sort()
print("Yeni liste(Sort): ", liste)
```

```python
# Sözlüklerde yerleşik metotlar

# Sözlük oluşturma
soz1 = {'ad': 'Hakan', 'yas': 30, 'sehir': 'Düzce'}

# get() belirtilen elemanı kullanarak değeri alır , eğer anahtar bulunmazsa belirtilen varsayılan değeri dödürür.
yas = soz1.get('yas', 'Anahtar bulunamadı')
print("Get : ", yas)

# items(): Sözlüğün öğelerini (anahtar-değer çiftleri) içeren görünüm nesnesini döndürür.
ogeler = soz1.items()
print("Items : ", ogeler)

# keys(): Sözlüğün anahtarlarını içeren bir görünüm nesnesini döndürür.
anahtarlar = soz1.keys()
print("Keys(): ", anahtarlar)

# values(): Sözlüğün değerlerini içeren bir görünüm nesnesini döndürür.
degerler = soz1.values()
print("Values(): ", degerler)

# copy(): Sözlüğün bir kopyasını oluşturur.
kopya_sozluk = soz1.copy()
print("Copy(): ", kopya_sozluk)

# setdefault(): Belirtilen anahtarın değerinin alır, eğer anahtar bulunmazsa nelirtilen varsayılan değeri ekler ve bu değeri döndürür.
sehir = soz1.setdefault('sehir', 'Bilinmiyor')
print("Setdefault(): ", sehir)

# update(): Yeni sözlükten veya anahtar-değer çiftlerinden gelen verilerle mevcut sözlüğü günceller.
yeni_veri = {'yas': 31, 'cinsiyet': 'Erkek'}
soz1.update(yeni_veri)
print('Update(): ', soz1)

# pop(): Belirtilen değerin anahtarın değerini çıkarır ve çıkarılan değeri döndürür.
pop_deger = soz1.pop('yas')
print('Pop(): ', pop_deger)

# popitem(): Sözlükte anahtar-değer çiftini çıkarır ve çıkarılan çifti döndürür.
cikarilan_ogeler = soz1.popitem()
print('Popitem(): ', cikarilan_ogeler)

# Güncel durumu yazdırır
print('Güncel: ', soz1)

# clear(): Sözlüğü temizler, tüm anahtar-değer çiftlerini kaldırır.
soz1.clear()
print("Clear:", soz1)
```

```python
# Kümelerde yerleşik metotlar

# İki küme oluşturalım

kume1 = {1, 2, 3, 4, 5}
kume2= {3, 4, 5, 6, 7}

# add: Bir küme elemanı ekler.
kume1.add(6)
print("Ekle: ", kume1)

# discard: Bir elemanı kümeden çıkarır (eğer kümede 1se)
kume1.discard(3)
print("Çıkart: ", kume1)

# difference: İki küme arasındaki farkı döndürür
fark_kume = kume1.difference(kume2)
print("Difference(): ", fark_kume)

# difference_update: Bir kümenin diğerine göre farkını günceller
kume1.difference_update(kume2)
print("Farkı Güncelle: ", kume1)

# intersection: İki kümenin kesişimini döndürür
kesisim_kume = kume1.intersection(kume2)
print("Kesişim: ", kesisim_kume)

# isdisjoint: İki kümenin kesişimi boş mu kontrol eder
kesisim_bos_mu = kume1.isdisjoint(kume2)
print("Kesişim boş mu :", kesisim_bos_mu)

# issubset: Bir küme, diğer kümenin alt kümesi mi kontrol eder
alt_kume_mı = kume1.issubset(kume2)
print("Alt küme mi : ", alt_kume_mı)

# union: İki kümenin birleşimini döndürür
birlesim_kume = kume1.union(kume2)
print("Birleşim küme: ", birlesim_kume)
```

```python
# Parametre alan fonksiyonlar

import math


def main():
  a =float(input(" a kenarını giriniz : "))
  b =float(input(" b kenarını giriniz : "))
  c =float(input(" c kenarını giriniz : "))
  alan  = Heron(a,b,c)
  print("Üçgenin alanı : ",alan)

def Heron(a,b,c):
  """
  Üç kenarı girilmiş üçgenin alanını hesaplar

  Parametres:
  a (float): Üçgenin 1. kenar uzunluğu
  b (float): Üçgenin 2. kenar uzunluğu
  c (float): Üçgenin 3. kenar uzunluğu

  Returns:
  alan (float): Üçgenin alanı.
  """

  u = (a+b+c)/2
  alan = math.sqrt(u * (u-a) * (u-b) * (u-c))
  return alan

main()

help(Heron)
```

```python
# Girilen 2 sayı arasındaki sayıların 1'lerle 100'ler  basamağı ile 10'lar ve 1000'ler basamağı arasında eşit olanları bize verir

def sayi_bul(ilk,son):
  for i in range(ilk,son+1,2):
    birler = i % 10
    onlar = (i//10) % 10
    yuzler = (i//100) % 10
    binler = (i//1000) % 10

    if (birler + yuzler == onlar + binler):
      print(i)

ilk = int(input("ilk sayıyı giriniz : "))
son = int(input("son sayıyı giriniz : "))

sayi_bul(ilk,son)
```

```python
def asal_sayi_kontrol(sayi):
  if sayi < 2:
    return False
  for i in range(2,int(sayi ** 0.5)+1):
    if sayi %i == 0:
      return False
  return True

def asal_sayilari_bul_ve_yazdir(limit):
  asal_sayilar = []
  for sayi in range(2,limit + 1):
    if asal_sayi_kontrol(sayi):
      asal_sayilar.append(sayi)
  print(f"1-{limit} arasındaki asal sayılar : ")
  for asal in asal_sayilar:
    print(asal)

limit = int(input("Bir sayı giriniz : "))
asal_sayilari_bul_ve_yazdir(limit)


```

```python
def ayir_listeler(string):
  sesli_harfler = []
  sessiz_harfler = []
  diger_karakterler = []

  sesli_harfler_turkce = 'aeıioöuüAEIİOÖUÜ'

  for karakter in string:
    if karakter.isalpha():
      # karakter harf ise,sesli veya sessiz kontrolü yapılır
      if karakter in sesli_harfler_turkce:
        sesli_harfler.append(karakter)
      else:
        sessiz_harfler.append(karakter)
    else:
      # karakter harf değilse, diğer karakter listesine ekler
      diger_karakterler.append(karakter)

  return sesli_harfler,sessiz_harfler,diger_karakterler


metin = input("Bir string giriniz :")

# Fonksiyonu çağırarak listelere ayrılmasını sağla
sesli, sessiz,diger = ayir_listeler(metin)


print("Sesli harfler : ",sesli)
print("Sessiz harfler : ",sessiz)
print("Diğer harfler : ",diger)


```

```python
def faktoriyel_hesaplama(sayi):
  if sayi == 0 or sayi ==1:
    return 1
  else:
    return sayi*faktoriyel_hesaplama(sayi-1)


def seri_hesaplama(n):
  toplam = 0.0 # serinin toplamı

  for i in range(1,n+1):
    toplam += i / faktoriyel_hesaplama(i)

  return toplam

n_terim = int(input("Terim sayısını giriniz : "))
sonuc = seri_hesaplama(n_terim)
print(f"1/1! + 2/2! + 3/3! + ... + {n_terim}/{n_terim}! serisinin değeri : {sonuc:.2f}")
```

```python
def urun_ekle():
  toplam_fiyat = 0
  while True:
    urun_adi = input(("ürün adını girin (çıkış için 'q):"))
    if urun_adi.lower() == 'q':
      break
    fiyat = float(input("ürün fiyatını girin.: "))
    if fiyat < 0 or fiyat > 1000:
      print("geçersiz fiyat fiyat 0 ile 1000 arasında olmalı.")
      continue
    toplam_fiyat += fiyat
    print(f"{urun_adi} ürünü sepete eklendi. toplam fiyat: {toplam_fiyat}")
  print("alışveriş tamamlandı. toplam ödeme:", toplam_fiyat)
urun_ekle()
```
